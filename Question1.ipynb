```python
from keras.datasets import fashion_mnist
import numpy as np
import pandas as pd
import wandb
import matplotlib.pyplot as plt
X,Y= fashion_mnist.load_data()
X_train=X[0]
Y_train=X[1]
X_test=Y[0]
Y_test=Y[1]
no_of_classes=10
mapping = {0: "T-shirt/top", 1: "Trouser", 2: "Pullover", 3: "Dress", 4: "Coat", 5: "Sandal", 6: "Shirt", 7: "Sneaker", 8: "Bag", 9: "Ankle boot"}
```

```python
wandb.login()
run = wandb.init(project='Assignment1')
```

```python
def sample_images(X_train,Y_train,mapping,no_of_classes):
  plt.figure()
  images=[]
  labels=[]
  i=0
  while i<no_of_classes:
    index= np.argmax(Y_train==i)
    image = X_train[index,:,:]
    plt.subplot(3,6, i+1)
    plt.imshow(image)
    plt.title(mapping[i])
    images.insert(i, image)
    labels.insert(i,mapping[i])
    i=i+1
  for i, j in zip(images, labels):
    wandb.log({"1": [wandb.Image(i, caption=j)]})












```

```python
sample_images(X_train,Y_train,mapping,no_of_classes)
```
